# Les dictionnaires ! 

## Lecture du csv (Genre, Espèce et Année de plantation)

```python
import csv

my_trees = []
with open('../arbres-grenoble/data/trees.csv', newline='') as csvfile:
    trees = csv.reader(csvfile, delimiter=',') 
    for row in trees:
        my_trees.append([row[13], row[14], row[19]]) 
        
```

## On regarde à quoi ça ressemble 

On a pas mal de valeurs manquantes...

```python
my_trees[0:6]
```

## Construction d'une liste "genre"

```python
genre = []
for arbre in my_trees:
    genre.append(arbre[0])
    
genre[0:6]
```

## Quels sont les éléments uniques dans 'genre' ? 

```python
genre_uniques = set(genre[1:])
genre_uniques.remove('')

genre_uniques
```

## Creation d'un dictionnaire de la forme {genre : nombre}

```python
# Un exemple pour visualiser ce qu'on a besoin de "faire boucler"
{'Abies':genre.count('Abies')}
```

```python
my_dict = {}
for key in genre_uniques:
    my_dict[key] = genre.count(key)  
    
my_dict
```

Dans l'exemple précedent key prend successivement les valeurs : 'Acer', 'Platanus', 'Fraxinus', ... 

my_dict[key] : val 
1. Si la clé 'key' existe déjà => remplace la valeur déjà présente dans la clé par la nouvelle valeur 'val'
2. Si la clé 'key' n'existe pas => créé la clé 'key' et lui donne la valeur val 

```python
# Nombre de Paulownia
my_dict['Paulownia']
```

## Creation d'un dictionnaire de la forme { (genre, espece) : nombre }
(genre, espece) en key, et le nombre de représentants en value

```python
# Création d'une liste de tuple (genre, espece)
list_of_tuple = []
for row in my_trees[1:]:
    if row[1] != '':
        list_of_tuple.append( tuple([row[0],row[1]]) )
    else:
        list_of_tuple.append( tuple([row[0],'unknown']))
```

```python
# On regarde  
list_of_tuple[:3]
```

```python
# Création du 'set' contenant toutes les tuples uniques  
dict_genre_espece = {}
tuples_uniques = set(list_of_tuple)

tuples_uniques
```

```python
# On se sert de ce set pour itérer et construire le dictionnaire 
for key in tuples_uniques: 
    dict_genre_espece[key]= list_of_tuple.count(key)
    
dict_genre_espece
```

## Deux exemples pour compter le nombre d'arbres plantés depuis 1984


```python
# Création de la liste contenant les années en utilisant une "liste comprehension"
l_annee_plantation = [row[2] for row in my_trees[1:]]

l_annee_plantation
```

### Creation d'un dictionnaire de la forme { annee_de_plantation : nombre_arbres_plantés }


```python
dict_year = {}
for key in set(l_annee_plantation):
    dict_year[key] = l_annee_plantation.count(key)

dict_year
```

```python
# Je fabrique la somme cumulative sur le dictionnaire
# remarquer la comparaison key > '1984' 
# => on peut comparer des choses comme ceci '1975' > '1950'
somme = 0
for key in dict_year.keys():
    if key > '1984':
          somme = somme + dict_year[key]
            
somme
```

```python
# Alternative à la cellule précédente en utilisant une 'list comprehension'
sum([val[1] for val in dict_year.items() if val[0]>'1984'])
```
